## Cell 1: Project Initialization & Dependencies
This cell ensures all necessary libraries from your requirements.txt are installed.

```python
import os
import subprocess
import sys

# 1. Install dependencies from the requirements file
if os.path.exists("../requirements.txt"):
    print("Installing dependencies from requirements.txt...")
    subprocess.check_call([sys.executable, "-m", "pip", "install", "-r", "../requirements.txt"])
    print("Dependencies installed successfully.")
else:
    print("requirements.txt not found. Skipping installation.")

import requests
import pandas as pd
import spacy

model_name = "en_core_web_sm"

if not spacy.util.is_package(model_name):
    print(f"Downloading {model_name}...")
    spacy.cli.download(model_name)
    print(f"{model_name} downloaded successfully.")
else:
    print(f"{model_name} is already installed.")
```

## Cell 2: Automated Directory Structure
This cell creates the folder hierarchy needed for the project. If the folders already exist, it skips creating them.

```python
# Define the project folder structure
folders = [
    '../data/raw',
    '../data/processed',
    '../outputs/models',
    '../outputs/embeddings',
    '../outputs/figures',
    '../outputs/rewrites',
    '../outputs/results',
]

print("Creating directory structure...")
for folder in folders:
    os.makedirs(folder, exist_ok=True)
    print(f"Verified: {folder}")

print("\nAll directories are ready.")
```

### Cell 3: Data Acquisition (Google Drive)
Downloads the **Persuade** dataset, main dataset from this project, from Google Drive and saves it to the `/data` directory.

```python
import gdown

file_id = '1aqMZMQIXqfQXnvRF8L8PeklgKoV2I6pG'
url = f'https://drive.google.com/uc?id={file_id}'
output_path = '../data/raw/persuade_2.csv'

if not os.path.exists(output_path):
    print("Downloading dataset from Google Drive...")
    gdown.download(url, output_path, quiet=False)
    print(f"Dataset saved to {output_path}")
else:
    print("Dataset already exists in ../data/raw/. Skipping download.")
```
